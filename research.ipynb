По первичному анализу можно сделать вывод, что в нашем датесете есть пропощуенные значения, которые имеют значение, а потому не требуют удаления, а требуют категоризации и последующей кодировки для которой можно использовать OneHotEncoder или LabelEncoder.

Также появились мысли о том, что было бы интересно рассмотреть корреляцию между признаками. Посмотреть, как влияет на количество просмотров реклама, насколько она хорошо воздействует на конверсию.

В следующих шагах перейдем к категоризации данных и их дальнейшей обработке.

```python
from sklearn.preprocessing import LabelEncoder

df.loc[df["тип рекламы"]=="c","тип рекламы"] = 1
df.loc[df["тип рекламы"].isna(),"тип рекламы"] = 0
df["тип рекламы"] = df["тип рекламы"].astype(int)

df.loc[df["рекламный слоган"] == "РАСПРОДАЖА","рекламный слоган"] = 1
df.loc[df["рекламный слоган"].isna(),"рекламный слоган"] = 0
df["рекламный слоган"] = df["рекламный слоган"].astype(int)

df.loc[df["участвует в продвижении?"] == 1.0,"участвует в продвижении?"] = 1
df.loc[df["участвует в продвижении?"].isna(),"участвует в продвижении?"] = 0

df.loc[df["ставка за тысячу показов"].isna(),"ставка за тысячу показов"] = 0
df.loc[df["место товара при продвижении"].isna(),"место товара при продвижении"] = df["место товара без продвижения"]


le = LabelEncoder()
df["код бренда"] = le.fit_transform(df["бренд"].astype(str))
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df_num = df.drop(columns=["название","бренд"])
df_num.info()
```

```python
import matplotlib.pyplot as plt
correlation_matrix = df_num.corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Матрица корреляций')

plt.xticks(ticks=range(len(correlation_matrix.columns)), labels=correlation_matrix.columns, rotation=90)
plt.yticks(ticks=range(len(correlation_matrix.columns)), labels=correlation_matrix.columns)
plt.show()
```

```python
filtered_correlation = correlation_matrix[abs(correlation_matrix)>0.2]
filtered_correlation
```

Таким образом, что нам дал первичный анализ:

1. Данные были обработаны в более красивый вид, а именно были обработаны пропущенные значения

2. Были выявлены интересные корреляции между признаками. Они позволили нам дать следующие выводы:
* есть небольшая корреляция между ценой и стоимостью доставки
* небольшая корреляция между рейтингов продавца и рейтингом товара
* между количеством отзывов и количеством картинок
* между количеством отзывов и количеством цветов
* между количеством картинок и рейтингом продавца
3. Также были подтверждены следующие корреляции между 
- местом товара без продвижения и участием в продвижении товара
